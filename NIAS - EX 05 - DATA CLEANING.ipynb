```python
import numpy as np
import pandas as pd

#importando as tabelas
test_csv = pd.read_csv('../input/titanic/test.csv')
train_csv = pd.read_csv('../input/titanic/train.csv')
```

**(1) ANÁLISE INICIAL**

```python
print('Colunas do BD de testes e treino(Temos apenas a coluna de Survived como um extra em relação com o BD testes): ')
print(train_csv.columns)
print()
print(test_csv.dtypes)
print()
print(train_csv.dtypes)


print('\nQuantidades de NaN em Test e em Train: ')

print(test_csv.isnull().sum())
print()
print(train_csv.isnull().sum())
```

**(2) LIDANDO COM VALORES NULOS**

```python
test_new = test_csv.drop(['Cabin', 'Ticket'], axis=1)
train_new = train_csv.drop(['Cabin', 'Ticket'], axis=1)

change=['Age', 'Fare']

test_new[change] = test_new[change].fillna(test_new[change].median())
train_new[change] = train_new[change].fillna(train_new[change].median())

test_new['Embarked'] = test_new['Embarked'].fillna(test_new['Embarked'].mode)
train_new['Embarked'] = train_new['Embarked'].fillna(train_new['Embarked'].mode)
```

**(3) CRIAÇÃO DE GRUPOS PARA FACILITAR A ANÁLISE**

```python
test_new['Age_Intervals'] = pd.cut(test_new['Age'], bins=5)
train_new['Age_Intervals'] = pd.cut(train_new['Age'], bins=5)

test_new['Fare_Intervals'] = pd.cut(test_new['Fare'], bins=6)
train_new['Fare_Intervals'] = pd.cut(train_new['Fare'], bins=6)
test_new
```
